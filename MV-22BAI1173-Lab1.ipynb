# Question 1
Perform Arithmetic operations (Addition, Subtraction, Multiplication, Division) and Logical operations (AND, OR, XOR) on images (color and grey scale images) using open CV and without open CV.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Load two images (color and grayscale)

```python
# Read the first image in color mode
img1_color = cv2.imread('image1.jpg')
# Read the second image in color mode
img2_color = cv2.imread('image2.jpg')
# Convert the first image to grayscale
img1_gray = cv2.cvtColor(img1_color, cv2.COLOR_BGR2GRAY)
# Convert the second image to grayscale
img2_gray = cv2.cvtColor(img2_color, cv2.COLOR_BGR2GRAY)
```

## Resize images to the same size for pixel-wise operations

```python
# Resize the first color image to 300x300 pixels
img1_color = cv2.resize(img1_color, (300, 300))
# Resize the second color image to 300x300 pixels
img2_color = cv2.resize(img2_color, (300, 300))
# Resize the first grayscale image to 300x300 pixels
img1_gray = cv2.resize(img1_gray, (300, 300))
# Resize the second grayscale image to 300x300 pixels
img2_gray = cv2.resize(img2_gray, (300, 300))
```

## Arithmetic Operations using OpenCV

```python
# Add the two color images
addition_cv = cv2.add(img1_color, img2_color)
# Subtract the second color image from the first
subtraction_cv = cv2.subtract(img1_color, img2_color)
# Multiply the two color images
multiplication_cv = cv2.multiply(img1_color, img2_color)
# Divide the first color image by the second
division_cv = cv2.divide(img1_color, img2_color)
```

```python
# Add the two gray images
addition_cv_gray = cv2.add(img1_gray, img2_gray)
# Subtract the second gray image from the first
subtraction_cv_gray = cv2.subtract(img1_gray, img2_gray)
# Multiply the two gray images
multiplication_cv_gray = cv2.multiply(img1_gray, img2_gray)
# Divide the first gray image by the second
division_cv_gray = cv2.divide(img1_gray, img2_gray)
```

## Logical Operations using OpenCV

```python
# Perform bitwise AND operation on the two color images
and_cv = cv2.bitwise_and(img1_color, img2_color)
# Perform bitwise OR operation on the two color images
or_cv = cv2.bitwise_or(img1_color, img2_color)
# Perform bitwise XOR operation on the two color images
xor_cv = cv2.bitwise_xor(img1_color, img2_color)
```

```python
# Perform bitwise AND operation on the two gray images
and_cv_gray = cv2.bitwise_and(img1_gray, img2_gray)
# Perform bitwise OR operation on the two gray images
or_cv_gray = cv2.bitwise_or(img1_gray, img2_gray)
# Perform bitwise XOR operation on the two gray images
xor_cv_gray = cv2.bitwise_xor(img1_gray, img2_gray)
```

## Arithmetic Operations without OpenCV (using NumPy)

```python
# Add the two color images using NumPy
addition_np = img1_color + img2_color
# Subtract the second color image from the first using NumPy
subtraction_np = img1_color - img2_color
# Multiply the two color images using NumPy
multiplication_np = img1_color * img2_color
# Divide the first color image by the second using integer division (avoid divide by zero)
division_np = img1_color // (img2_color + 1)  # Adding 1 to avoid division by zero
```

```python
# Add the two gray images using NumPy
addition_np_gray = img1_gray + img2_gray
# Subtract the second gray image from the first using NumPy
subtraction_np_gray = img1_gray - img2_gray
# Multiply the two gray images using NumPy
multiplication_np_gray = img1_gray * img2_gray
# Divide the first gray image by the second using integer division (avoid divide by zero)
division_np_gray = img1_gray // (img2_gray + 1)  # Adding 1 to avoid division by zero
```

## Logical Operations without OpenCV (using NumPy)

```python
# Perform bitwise AND operation using NumPy
and_np = np.bitwise_and(img1_color, img2_color)
# Perform bitwise OR operation using NumPy
or_np = np.bitwise_or(img1_color, img2_color)
# Perform bitwise XOR operation using NumPy
xor_np = np.bitwise_xor(img1_color, img2_color)
```

# Question 2
Geometric Transformations
- Crop a specific region from an image.

  a. Input: A sample image and crop coordinates.

  b. Output: Display the cropped portion of the image.

- Warp an image using affine transformation.

  a. Input: A sample image and transformation matrix.

  b. Output: Perform and display the warped image.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Load a sample image

```python
# Read the image in color mode
img_color = cv2.imread('image1.jpg')
# Convert the image to grayscale
gray_img = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
```

## =================== CROP A SPECIFIC REGION ===================

```python
# Define crop coordinates (x, y, width, height)
crop_x, crop_y, crop_w, crop_h = 50, 50, 200, 200
```

```python
# Crop the region from the color image
cropped_color = img_color[crop_y:crop_y+crop_h, crop_x:crop_x+crop_w]
# Crop the region from the grayscale image
cropped_gray = gray_img[crop_y:crop_y+crop_h, crop_x:crop_x+crop_w]
```

```python
# Display the cropped images
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(cropped_color, cv2.COLOR_BGR2RGB))
plt.title('Cropped Color Image')

plt.subplot(1, 2, 2)
plt.imshow(cropped_gray, cmap='gray')
plt.title('Cropped Grayscale Image')

plt.tight_layout()
plt.show()
```

## =================== AFFINE TRANSFORMATION ===================

```python
# Define three points for the affine transformation
rows, cols, _ = img_color.shape
```

### =================== TRANSLATION ===================

```python
# Define translation matrix: Move image by (50, 50)
M_translation = np.float32([[1, 0, 50], [0, 1, 50]])
# Apply the translation to the color image
translated_color = cv2.warpAffine(img_color, M_translation, (cols, rows))
# Apply the translation to the grayscale image
translated_gray = cv2.warpAffine(gray_img, M_translation, (cols, rows))
```

### =================== SCALING ===================

```python
# Define scaling matrix: Scale image by 1.5 times
M_scaling = np.float32([[1.5, 0, 0], [0, 1.5, 0]])
# Apply the scaling to the color image
scaled_color = cv2.warpAffine(img_color, M_scaling, (int(cols * 1.5), int(rows * 1.5)))
# Apply the scaling to the grayscale image
scaled_gray = cv2.warpAffine(gray_img, M_scaling, (int(cols * 1.5), int(rows * 1.5)))
```

### =================== SHEARING ===================

```python
# Define shearing matrix: Shear image horizontally
M_shearing = np.float32([[1, 0.5, 0], [0.5, 1, 0]])
# Apply the shearing to the color image
sheared_color = cv2.warpAffine(img_color, M_shearing, (cols + 100, rows + 100))
# Apply the shearing to the grayscale image
sheared_gray = cv2.warpAffine(gray_img, M_shearing, (cols + 100, rows + 100))
```

```python
# Display the transformed images
plt.figure(figsize=(12, 8))

# Display translated images
plt.subplot(3, 2, 1)
plt.imshow(cv2.cvtColor(translated_color, cv2.COLOR_BGR2RGB))
plt.title('Translated Color Image')

plt.subplot(3, 2, 2)
plt.imshow(translated_gray, cmap='gray')
plt.title('Translated Grayscale Image')

# Display scaled images
plt.subplot(3, 2, 3)
plt.imshow(cv2.cvtColor(scaled_color, cv2.COLOR_BGR2RGB))
plt.title('Scaled Color Image')

plt.subplot(3, 2, 4)
plt.imshow(scaled_gray, cmap='gray')
plt.title('Scaled Grayscale Image')

# Display sheared images
plt.subplot(3, 2, 5)
plt.imshow(cv2.cvtColor(sheared_color, cv2.COLOR_BGR2RGB))
plt.title('Sheared Color Image')

plt.subplot(3, 2, 6)
plt.imshow(sheared_gray, cmap='gray')
plt.title('Sheared Grayscale Image')

plt.tight_layout()
plt.show()
```

# Question 3
Histogram Analysis

- Compute and plot the histogram of a grayscale image.

- Perform Histogram Equalization on an image.

  a. Input: A low-contrast grayscale image.

  b. Output: Show the improved image after equalization.

```python
import cv2
from matplotlib import pyplot as plt
```

## Load a sample image

```python
# Read the image in color mode
img_color = cv2.imread('image1.jpg')
# Convert the image to grayscale
gray_img = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
```

## =================== HISTOGRAM ANALYSIS ===================

```python
# Compute and plot the histogram of a grayscale image
plt.figure(figsize=(8, 4))
```

```python
# Compute histogram using OpenCV (bins: 256, range: [0, 256])
hist = cv2.calcHist([gray_img], [0], None, [256], [0, 256])
```

```python
# Plot the histogram
plt.subplot(1, 2, 1)
plt.plot(hist, color='black')
plt.title('Histogram of Grayscale Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
```

## Perform Histogram Equalization

```python
# Apply equalization to the grayscale image
equalized_img = cv2.equalizeHist(gray_img)
```

```python
# Display the equalized image
plt.subplot(1, 2, 2)
plt.imshow(equalized_img, cmap='gray')
plt.title('Equalized Grayscale Image')

plt.tight_layout()
plt.show()
```
